```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Make a GET request to the website URL
url = 'https://laverne.com/'
response = requests.get(url)

```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
products = soup.find_all('div', class_='product-entry__content')
products
```

```python
# Create a list to store all the scraped data
all_data = []
```

```python

# Loop through each product and extract the relevant data
for product in products:
    # Get the product name
    name = product.find('h3', class_='product-entry__title mt-0').text.strip()

    # Get the product price
    price = product.find('h4', class_='text-red-400 font-bold text-lg')

    # Get the product discount, if available
    discount_tag = product.find('span', class_='text-sm text-gray-400 line-through')
    if discount_tag:
        discount = discount_tag.text.strip()
    else:
        discount = ''

    # Add the data to the all_data list
    all_data.append([name, price, discount])
```

```python
# Write the data to a CSV file
with open('laverne_products.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Name', 'Price', 'Discount'])
    writer.writerows(all_data)
```
